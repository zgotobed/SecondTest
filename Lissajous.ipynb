```python
import numpy as np
import matplotlib.pyplot as plt

def plot_lissajous(a, b, delta, num_points):
    t = np.linspace(0, 2*np.pi, num_points)
    x = np.sin(a * t + delta)
    y = np.sin(b * t)
    
    plt.plot(x, y)
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.title(f'Lissajous Curve (a={a}, b={b}, delta={delta})')
    plt.grid(True)
    plt.show()

# Example usage
a = 3
b = 2
delta = np.pi/2
num_points = 1000

plot_lissajous(a, b, delta, num_points)
```
